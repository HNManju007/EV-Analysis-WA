```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs("../visuals", exist_ok=True)
```

```python
df = pd.read_csv(r"M:\EV_analysis\data\clean_ev_data.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
df['MAKE'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 EV Manufacturers")
plt.xlabel("Make")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r"M:\EV_analysis\visuals\Top 10 EV Manufacturers.png")
#plt.savefig("../visuals/top_ev_makes.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ELECTRIC_RANGE'], bins=30, kde=True, color='green')
plt.title("Distribution of Electric Vehicle Ranges")
plt.xlabel("Electric Range (miles)")
plt.ylabel("COUNT")
plt.tight_layout()
plt.savefig(r"M:\EV_analysis\visuals\electric_range_distribution.png")
#plt.savefig("../visuals/electric_range_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['BASE_MSRP'])
plt.title("MSRP Price Distribution of EVs")
plt.xlabel("Base MSRP ($)")
plt.tight_layout()
#plt.savefig("../visuals/msrp_boxplot.png")
plt.savefig(r"M:\EV_analysis\visuals\msrp_boxplot.png")
plt.show()
```

```python
top_counties = df['COUNTY'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_counties.plot(kind='barh', color='orange')
plt.title("Top 10 Counties by EV Adoption")
plt.xlabel("Number of Registered EVs")
plt.ylabel("County")
plt.gca().invert_yaxis()
plt.tight_layout()
#plt.savefig("M:\EV_analysis\visuals"/top_counties_ev_adoption.png")
plt.savefig(r"M:\EV_analysis\visuals\top_counties_ev_adoption.png")

plt.show()
```
