```python
def clean_text_columns(df, cols):
    for col in cols:
        df[col] = df[col].astype(str).str.upper().str.strip()
    return df

# Use it like this:
text_cols = ['VIN_110', 'County', 'City', 'State', 'Make', 'Model', 
             'Electric_Vehicle_Type', 
             'Clean_Alternative_Fuel_Vehicle_CAFV_Eligibility', 
             'Vehicle_Location', 'Electric_Utility']

df = clean_text_columns(df, text_cols)
```
